```python
###  to design the bank systems --important
class Bank:
  def __init__(self,owner,money,password):
    self.owner=owner
    self.money=money
    self.password=password

  def deposit(self,amount=0):
    self.money+=amount
    print(f"{self.owner}! your final balance: {self.money}")

  def withdraw(self,amount=0):
    if amount>1000:
      print(f"Hey! withdrawing more money")
      n=int(input("Enter your password: "))
      if n==792101:
        self.money-=amount
        print(f"{self.owner}! your final balance: {self.money}")
      else:
        print("Wrong password")

obj1=Bank("Akash",20500,792101)   ## gives the address of the parameter where it stored
obj1.deposit(1001)
obj1.withdraw(1030)
```

```python
## to lower the value in string--important
class st:
  def __init__(self,value):
    self.value=value

  def lower_value(self):
    code=ord(self.value)
    if code>=65 and code<=90:
      new_code=code+32
      self.value=chr(new_code)
      print(self.value)
    else:
      print("Already in uppper case")

obj3=st("H")
obj3.lower_value()
```

```python
### example of inheritance --
class old:
  def __init__(self,name,money):
    self.name=name
    self.money=money
  def money_hel(self):
    print("hello how are you")

class new(old):
  def __init__(self, name, money,rollno):
    super().__init__(name, money)    ## it is used to take the inherent attributes of the parent class
    self.rollno=rollno       
  
  def pr(self):
    print(f"Your details: {self.name} {self.rollno}")

b1=new("akash",657900,2301100)
b1.money_hel()
print("\n")
```
